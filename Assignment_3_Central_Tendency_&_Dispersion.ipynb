# 3. Who scores the most on average for math,reading and writing based on Gender & Test preparation course?

# 4. What do you think about the scoring variation for math,reading and writing based on Gender & Test preparation course?

```python
math_var=stat.variance(data['math score'])
math_stdev=stat.stdev(data['math score'])
reading_var=stat.variance(data['reading score'])
reading_stdev=stat.stdev(data['reading score'])
writing_var=stat.variance(data['writing score'])
writing_stdev=stat.stdev(data['writing score'])
```

```python
print(math_var,math_stdev)
```

```python
print(reading_var,reading_stdev)
```

```python
print(writing_var,writing_stdev)
```

```python
gender_var=data[['gender','math score','reading score','writing score']].groupby('gender').std()
gender_var
```

```python
test_var=data[['test preparation course','math score','reading score','writing score']].groupby('test preparation course').std()
test_var
```
